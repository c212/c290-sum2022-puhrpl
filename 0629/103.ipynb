##Entanglement

In this notebook we will discuss a fundamental quantum
phenomenon that did not show up in the double slit experiment.
So this is called entanglement. And it's one of the basic features 
of quantum mechanics that's exploited in quantum computation. So let's 
imagine that we have a system of two qubits, say two hydrogen atoms.
And let's say that our first qubit is in this state

$$\alpha_0|0\rangle + \alpha_1|1\rangle$$

And the second qubit is in this state

$$\beta_0|0\rangle + \beta_1|1\rangle$$

Now we could ask, what's the state of the two qubits together as a composite system? So remember the state of the composite system is a superposition of where
there are four classical possibilities.
So it's of the form 

$$a_{00}|00\rangle + a_{01}|01\rangle + a_{10}|10\rangle + a_{11}|11\rangle$$


So now we could ask the question, how do you go from the description of the two qubits individually to a description of the composite system? Answer is: simple,  what we do is multiply these two representations together. Now, if you want to be precise about it, what we are actually doing is a kind of multiplication called taking a tensor product. But we'll see this in later lectures. So for the moment, just think of it naively as multiplying the two representations together. And so in the answer $a_{00} = \alpha_0 \beta_0$ and similarly
$a_{01} = \alpha_0\beta_1$ and so on (so, in general, $\alpha_i\beta_j=a_{ij})$. So just to make this clear we could  work out an example.


So let's say that our first qubit was in the state
$\frac{1}{\sqrt{2}}|0\rangle+\frac{1}{\sqrt{2}}|1\rangle$
and our second qubit was in the state 
$\frac{1}{2}|0\rangle-\frac{\sqrt{3}}{2}|1\rangle$. So now the composite state of the system is just 

$$\Big(\frac{1}{\sqrt{2}}|0\rangle+\frac{1}{\sqrt{2}}|1\rangle\Big) \Big(\frac{1}{2}|0\rangle-\frac{\sqrt{3}}{2}|1\rangle\Big) = \\
= \frac{1}{2\sqrt{2}}|00\rangle - \frac{\sqrt{3}}{2\sqrt{2}}|01\rangle + \frac{1}{2\sqrt{2}}|10\rangle - \frac{\sqrt{3}}{2\sqrt{2}}|11\rangle$$

Now you could ask a different question. You could say, well, suppose I give you this state of the composite system. How would you figure out the other direction?
How can you figure out the state of the two qubits,  individually? Well, if I gave you this particular example, then you could go back by just saying, that's what it is. You just factor this product into the individual components. And you say, that's the state of the first qubit, and that's the state of the second qubit.

Now, you might imagine that factoring in this way might sometimes be difficult. But there's no reason to believe that you can't do this in general. As we'll see soon, there are certain states for which you cannot do this factoring. So there are some states for which you can factor them, and there are other states for which you cannot do this factoring at all. And you might wonder, if you cannot do this, if you cannot do such
factorization, what does this mean about the state of the two qubits individually? The answer is going to turn out that for states where you cannot do this
factorization, you cannot talk about the state of each qubit by itself. The two qubits are in some entangled state. So let's see an example of this. And here's a famous example of this. It's called a Bell state.

$$|\psi\rangle = \frac{1}{\sqrt{2}}|00\rangle+\frac{1}{\sqrt{2}}|11\rangle$$

It's a state of the two qubits where the composite state is an equal superposition of $|00\rangle$, and $|11\rangle$. So now, let's try to factor this state, and  we try to write it as follows

$$\big(\alpha_0|0\rangle + \alpha_1|1\rangle\big)\big(\beta_0|0\rangle + \beta_1|1\rangle\big)$$

So now what happens? Well, the product is just 

$$\alpha_0\beta_0|00\rangle +
\alpha_0\beta_1|01\rangle +
\alpha_1\beta_0|10\rangle +
\alpha_1\beta_1|11\rangle $$

OK. So now comparing coefficients, we realize that 
two of these coefficients must be zero, 
specifically we need to have
$$\alpha_0\beta_1 = 0 \textrm{ and } \alpha_1\beta_0 = 0$$
But (as we compare coefficients) we also have that 
$$\alpha_0\beta_0 = \frac{1}{\sqrt{2}} \textrm{ and } \alpha_1\beta_1 = \frac{1}{\sqrt{2}}$$
And I claim that this is impossible.
You see, because the fact that $\alpha_0\beta_0$ and $\alpha_1\beta_1$ are non-zero means that $\alpha_0$, $\beta_0$, $\alpha_1$, $\beta_1 \ne 0$, they're all non-zero. But then that's not compatible with $\alpha_0\cdot\beta_1$ being $0$. So one of these two(either $\alpha_0$ or $\beta_1$) has to be 0, meaning one of these two (either $\alpha_0\beta_0$ or $\alpha_1\beta_1$) has to be 0.
And likewise if $\alpha_1\beta_0 = 0$ then again, one of either $\alpha_0\beta_0$ or $\alpha_1\beta_1$ has to be 0. So this is a contradiction. It's impossible. And so this is a state of two qubits which is incompatible with any state of the single qubit. You just cannot describe this state as a state of the first qubit and the state of the second qubit. The two qubits are inherently entangled with each other.

OK. So now let's try to understand a little more about
this entangled state. So let's say that we were to measure the state. So what would we see? The answer is, we'd see $|00\rangle$ with probability $\frac{1}{2}$ and $|11\rangle$ with probability $\frac{1}{2}$. 

$$P[00] = \frac{1}{2} \textrm{ and } P[11]=\frac{1}{2}$$

OK, but now remember that this is a state of two qubits. So we could also just measure the first qubit. Now what do we see? Well, the probability that we see $0$ is exactly $\frac{1}{2}$. But in case we see $0$, the new state of the two qubits is the part of this superposition that's compatible with the outcome (of having seen $0$). And so what we have to do is cross off the $|11\rangle$ in the formula, and the new state would
just be $|00\rangle$. Similarly, we'd see $1$ with probability $\frac{1}{2}$, and now our new state would be equal to $|11\rangle$. A similar thing holds if we measure the second qubit, and not the first. So the probability we see $0$ here (on the second qubit) would be $\frac{1}{2}$, the probability we see $1$ here (on the second qubit) would be $\frac{1}{2}$ and the new state would again just like before ($|00\rangle$ and $|11\rangle$ respectively, see arrows in the picture below). 

OK, but now, let's imagine that we put these two qubits in this Bell state. And what we do is we take one of these qubits, and we put it on a spaceship that's heading for Mars. And then with our qubit here on Earth and the other qubit on Mars, we repeat the same experiment. So now what happens? Well, the first qubit, when we measure it, we see $0$ and $1$ with equal probability. But as soon as we make the measurement, what quantum mechanics tells us is that the new state must be $|00\rangle$, or $|11\rangle$, depending on the outcome, which means that even though the second qubit is far away, if you were to measure it, we'd see the same outcome, whatever we have got (measured) here. So it turns out that entanglement is really a fundamental phenomenon. So for example, if you look at the two electrons in a covalent bond, then it turns out that these electrons have an internal state, which is a qubit. It's called spin. And this spin state can be a superposition. The spin of an electron is described as being either in the state up or in the state down. We'll talk about spin much later in the course. But for now, just think of it as a qubit which is in one of these two states. Now, when you have a covalent bond, the two electrons in this covalent bond are actually in this kind of a superposition

$$\frac{1}{\sqrt{2}}|\uparrow\downarrow\rangle -
\frac{1}{\sqrt{2}}|\downarrow\uparrow\rangle$$

So they are in this highly entangled state, this is like the state

$$\frac{1}{\sqrt{2}}|01\rangle - \frac{1}{\sqrt{2}}|10\rangle$$

The importance of this concept of entanglement was actually recognized quite early in the history of quantum mechanics. And here's a beautiful quote from Erwin Schrodinger from 1935 where he says, "I would not call entanglement one but rather the characteristic trait of quantum mechanics, the one that enforces its entire departure from classical lines of thought." So we'll talk a lot more about entanglement and its properties in the rest of this lecture, and then in the next lecture.
