##Entanglement and Bell's Experiment

### K-Level Systems, Bra-Ket Notation

In this lecture, we'll talk about systems of two qubits. These quantum systems exhibit a remarkable property called entanglement. And this property of entanglement plays a critical role in quantum computation. So in this and the next couple of lectures, we'll try to get an intuition for entanglement and what this property's all about. But before we do that, in this video, I want to formalize what we spoke about last time about qubits and superpositions. So recall that the energy of an electron in an atom is quantized. What this means is that the electron cannot take on any arbitrary energy, but the energy must be in one of a discrete number of energy levels. So now imagine that we've limited the energy of the electron so that it's in the ground state, or the first excited state, and so on, up through the $(k-1)^\textrm{st}$ excited state. So if this electron were a classical system, then it could store one of $k$ pieces of information, which we might denote by $|0\rangle$, $|1\rangle$, through $|k-1\rangle$. Now the superposition principle, which is one of the basic axioms of quantum mechanics says that, in general, the state of the system is going to be any linear superposition of these allowable states.

So what the superposition principle says is that the general state is a
linear superposition of $|0\rangle$ through $|k-1\rangle$, each with its amplitude, $\alpha_j$, which is a complex number, and these amplitudes are normalized so that summation of $\alpha_j$ magnitude squared is equal to $1$ for $j$ equal to $0$ to $k-1$:

$$|\psi\rangle = \alpha_0|0\rangle + \alpha_1|1\rangle + ... + \alpha_{k-1}|k-1\rangle,~\textrm{with}~\alpha_j \in \mathbb{C},~\textrm{and such that}~\sum_{j=0}^{k-1}|\alpha_j|^2 = 1$$

And as we saw last time, interpreting what the state means is not very easy, 
because it's hard to make sense of what it means when we say that the electron is in the ground state with amplitude $-\frac{1}{2}$ or $\big(\frac{1}{2} + \frac{i}{2}\big)$. Of course, one way we can interpret this is through the
measurement axiom, which says that when we measure the system (or when we
look at it) then the probability that the outcome is $|j\rangle$ is the magnitude of $\alpha_j^2$, and the fact that the state is normalized means that we get, with probability $1$, to see (when we measure) some outcome $|j\rangle$ between $|0\rangle$ and $|k - 1\rangle$. Moreover, we also saw that the measurement disturbs the system and that the new state, which we'll denote by $|\psi'\rangle$, is the $j^\textrm{th}$ excited state if the measurement outcome was $|j\rangle$.

So if you want now we can do a quick example and if $k$ was $3$, and we had a three-state system, we might have our state as $|0\rangle$ with amplitude $\big(\frac{1}{2}+\frac{i}{2}\big)$, $|1\rangle$ with amplitude $-\frac{1}{2}$, and $|2\rangle$ with amplitude $\frac{i}{2}$:

$$|\psi\rangle = \Big(\frac{1}{2}+\frac{i}{2}\Big)|0\rangle - \frac{1}{2} |1\rangle + \frac{i}{2} |2\rangle$$

And now if we measure, the probability that we see $0$ is $P[0] = \frac{1}{4} + \frac{1}{4} = \frac{1}{2}$, and the new state is $|\psi'\rangle = |0\rangle$. The probability that we see $1$ is $P[1] = \frac{1}{4}$, and the new state is $|\psi'\rangle = |1\rangle$. And the probability we see $2$ is $P[2] = \frac{1}{4}$, and the new state would be just $|\psi'\rangle = |2\rangle$ (or, as we say, ket 2).


So now, let's look at a geometric interpretation of the quantum state. So geometrically, here's what the superposition principle says: it says that the state of a quantum system, a $k$-level quantum system, is a unit vector in a $k$-dimensional complex vector space. So the state is given by a unit vector in a k-dimensional complex vector space, and this vector space is also called a Hilbert space. What do we mean by this? Well, like we said before, we have a $k$-dimensional complex vector space, which has an orthonormal basis consisting of the states $|0\rangle$, $|1\rangle$, $|2\rangle$, through $|k-1\rangle$, and the state $|\psi\rangle$ is a unit vector. 

And if we write for our 3-level system

$$|\psi\rangle =  \alpha_0|0\rangle + \alpha_1 |1\rangle + \alpha_2|2\rangle$$

We could also have written it as (in standard vector notation)

$$|\psi\rangle = \alpha_0|0\rangle + \alpha_1 |1\rangle + \alpha_2|2\rangle = \begin{pmatrix} \alpha_0 \\ \alpha_1 \\ \alpha_2 \end{pmatrix} $$

And $\alpha_0$, $\alpha_1$, $\alpha_2$ are complex numbers ($\alpha_0, \alpha_1, \alpha_2 \in \mathbb{C}$).

So it (the state) belongs to a three-dimensional complex vector space.

We can also talk about what happens when we measure the system.
And we know that when we measure the system the state vector $|\psi\rangle$ gets projected onto one of the basis states. So if you're measuring in the standard basis {$|0\rangle$, $|1\rangle$, $|2\rangle$}, then $|\psi\rangle$ gets projected onto the state $|0\rangle$ with probability which is cosine squared of the angle $\theta_0$ it makes with the state $|0\rangle$. So the probability that the outcome is zero is $\cos^2\theta_0$. And if that's the outcome, then $|\psi\rangle$ is projected onto the zero state, and it actually becomes $|0\rangle$. 

Now this is true, in general, that the state vector is projected onto each of
the vectors in the orthonormal bases in which we are measuring, with
probability $\cos^2\theta$, where $\theta$ is the angle which it
makes with that particular vector. But how do we define $\theta$ in general?
So if you are given two different vectors, and maybe they are complex
vectors, how to we define $\theta$ (or $\cos{\theta}$) that's the question that we would like to address now. 

So assume two vectors: 

$$|\psi\rangle = \begin{pmatrix} \alpha_0 \\ \alpha_1 \\ \alpha_2 \end{pmatrix}\textrm{ and }|\phi\rangle = \begin{pmatrix} \beta_0 \\ \beta_1 \\ \beta_2 \end{pmatrix}\textrm{ and }\alpha_1, \alpha_2, \alpha_2, \beta_0, \beta_1, \beta_2 \in \mathbb{C}$$

So if you have two vectors in a 3-dimensional complex space then the way we define the cosine of the angle between them, is via the inner product between $|\psi\rangle$ and $|\phi\rangle$ which you get by taking the sum of the corresponding products (and if these are complex numbers, then you take convex
conjugates of one side). So the inner product is just 

$$\langle \psi | \phi \rangle = \bar\alpha_0\beta + \bar\alpha_1\beta_1 + \bar\alpha_2\beta_2$$

Which is also given by taking the row vector with complex conjugate (elements) of $|psi\rangle$, and then the column vector, $|\phi\rangle$.

$$\langle \psi | \phi \rangle = 
\begin{pmatrix} \bar\alpha_0 & \bar\alpha_1 & \bar\alpha_2 \end{pmatrix} \cdot 
\begin{pmatrix} \beta_0 \\ \beta_1 \\ \beta_2 \end{pmatrix}
$$

And we can define the angle between these two vectors, $\theta$,  by saying that
$\cos{\theta}$ is the absolute value of the inner product. And if these are complex numbers, then it's not the absolute value. It's the magnitude of the complex number: 

$$\cos{\theta} = | \bar\alpha_0\beta + \bar\alpha_1\beta_1 + \bar\alpha_2\beta_2 | = | \langle \psi | \phi \rangle |$$

###Example: measurement in the sign basis

So let's say that we're working in a two-dimensional space.
So here is our standard basis: $|0\rangle$ and $|1\rangle$ (shown).
So we are working with a qubit $|\psi\rangle$. These are the standard basis states, zero and one. These are the states plus and minus that we talked about last time. $|+\rangle$ is an equal superposition of zero and one.
$|-\rangle$ is the state which is $\frac{1}{\sqrt{2}}|0\rangle - \frac{1}{\sqrt{2}}|1\rangle$. And now, let's take a state $|\psi\rangle$, which looks like this. It's $|\psi\rangle = \frac{1}{2}|0\rangle + \frac{\sqrt{3}}{2}|1\rangle$

So now you know that $|+\rangle$, of course, makes a 
$45^\circ$ angle with $|0\rangle$. And what about $|\psi\rangle$? It 
makes a $60^\circ$ angle. And so the angle between $|\psi\rangle$ and 
$|+\rangle$ is $15^\circ$. So if you were measuring $|\psi\rangle$ in the plus-minus basis, the probability you would see plus is, of course, cosine squared 15. But how would you work this out as a vector? Well what you would say is that the probability that the outcome of the measurement is plus is just given by the inner product between these two vectors. So what's the inner product? Well you can just multiply corresponding coordinates. Or you can write it like 

$$P[+] = \Bigg[ \begin{pmatrix}\frac{1}{2} & \frac{\sqrt{3}}{2} \end{pmatrix} \cdot \begin{pmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}}\end{pmatrix} \Bigg]^2 = \bigg(\frac{1}{2\sqrt{2}} + \frac{\sqrt{3}}{2 \sqrt{2}}\bigg)^2 = \\
=  \bigg(\frac{1 + \sqrt{3}}{2\sqrt{2}}\bigg)^2 = \frac{1 + 3 + 2\sqrt{3}}{8} = \frac{2 + \sqrt{3}}{4}$$

Now of course, you can work out something similar for
probability of minus. But the other way you can do this 
is, of course you can realize that probability of plus 
plus probability of minus is 1, so the probability of minus 
is just 2 minus square root 3 over 4: 

$$P[-] = 1 - P[+] = 1 - \bigg( \frac{2 + \sqrt{3}}{4} \bigg) = \frac{4 - 2 - \sqrt{3}}{4} = \frac{2 - \sqrt{3}}{4}$$

Now let me show you one way you can do this calculation which, of course,
should give you the same value. But it might be worth doing this other calculation just to give you intuition. OK, so here's what we're going to do.
What we'll do is, we want to measure psi in the plus minus basis.
So what we'll do is we'll first rewrite psi as 

$$|\psi\rangle = \alpha |+\rangle + \beta |-\rangle$$

And now, once we write it like this, what's the probability of seeing plus 
when we measur it in the plus minus basis? It's just $|\alpha|^2$ (alpha magnitude squared). So how do you rewrite this in the plus minus basis? Well let's rewrite zero as a linear combination of plus and minus. And you can see here that $|0\rangle = \frac{1}{\sqrt{2}}|+\rangle + \frac{1}{\sqrt{2}} |-\rangle$ because if you add plus and minus, the $|1\rangle$ vector cancels and you get a multiple of zero. Similarly, $|1\rangle = \frac{1}{\sqrt{2}}|+\rangle - \frac{1}{\sqrt{2}}|-\rangle$. So from here we write $|\psi\rangle$ as follows: 

$$|\psi\rangle = \frac{1}{2} \bigg(\frac{1}{\sqrt{2}}|+\rangle + \frac{1}{\sqrt{2}} |-\rangle\bigg) + \frac{\sqrt{3}}{2} \bigg(\frac{1}{\sqrt{2}}|+\rangle - \frac{1}{\sqrt{2}} |-\rangle\bigg)$$

Which is, we can gather terms now, and we have: 

$$|\psi\rangle = \Bigg(\frac{1+\sqrt{3}}{2\sqrt{2}}\Bigg) |+\rangle + 
\Bigg(\frac{1+\sqrt{3}}{2\sqrt{2}}\Bigg) |-\rangle$$

And so if you were to figure out what's the probability that we'll see
plus when we do a measurement, we just take the square of this first complex amplitude. Which is something you'd recognize from the previous calculation.
So the probability of plus is just the square of this, which is 

$$P[+] = \Bigg(\frac{1+\sqrt{3}}{2\sqrt{2}}\Bigg)^2 = \frac{1 + 3 + 2\sqrt{3}}{8} = \frac{2 + \sqrt{3}}{4}$$

Which is what we saw before.
