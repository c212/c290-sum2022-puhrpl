#Two Qubits

OK so now we're going to talk about systems of two qubits. So continuing with our running example, imagine that we have a hydrogen atom. We want to use the ground or excited state of the electron to represent a bit of information. So now, since there are two such electrons, we can represent two bits of information classically. So there are four possible states of the two electrons: (0, 0), (0, 1), (1, 0), (1, 1).
So now, again, by the superposition principle, the quantum state of these
two electrons is in general a superposition of all four
possibilities.
So we can write 

$$| \psi \rangle = \alpha_{00}|00\rangle + \alpha_{01}|01\rangle + \alpha_{10}|10\rangle + \alpha_{11}|11\rangle$$

So what this is saying, basically, is that the two electrons don't make up their minds which of these four classical states they're in. And each of these four classical states has its own private amplitude, a complex number. 

So $\alpha_x \in \mathbb{C},~\forall x \in \{00, 01, 10, 11\}$ so these are complex numbers for each $x$ among these four
possibilities and $\sum|\alpha_x|^2=1$ so if you add together, the sum of the magnitudes squared of these four
amplitudes, you get 1.
So it's a normalized state.

Now again, what happens if you measure the state of these two qubits? OK, so if you look at these two qubits, they quickly
make up their minds--the electrons quickly make up their minds whether they're in ground or excited state. And so (if you measure) you end up seeing $|00\rangle$ with probability exactly $P[00] = |\alpha_{00}|^2$. And once again, the very act of making the measurement disturbs the system. And so the new state, $|\psi'\rangle$, would just be $|00\rangle$.
Similarly for the other four possibilities, the probability 
of $|01\rangle$ would be $P[01] = |\alpha_{01}|^2$ and  so on.
So for example, let's say that the state of our system was

$$|\psi\rangle = \Big(\frac{1}{2}+\frac{i}{2}\Big)|00\rangle + \frac{1}{2}|01\rangle - \frac{i}{2}|11\rangle$$

So now, the probability that we'll see a $|00\rangle$ would be exactly $P[00] = \frac{1}{2}$, the probability we'll see $|01\rangle$ would be $P[01] = \frac{1}{4}$ and the probability we'll see $|11\rangle$ would also be $P[11] = \frac{1}{4}$. 
OK, but now, we could ask a different question. 

Suppose we have the same qubit state. So suppose we had the state of the two qubits as: 

$$|\psi\rangle = \Big(\frac{1}{2}+\frac{i}{2}\Big)|00\rangle + \frac{1}{2}|01\rangle + \frac{i}{2}|11\rangle$$

Now, what happens if we were to measure only the first qubit?
What would we see? So let me do this in full generality, and then we'll work through the example. So let's say that in full  generality, this is our state.

$$|\psi\rangle = \alpha_{00}|00\rangle + \alpha_{01}|01\rangle +\alpha_{10}|10\rangle +\alpha_{11}|11\rangle$$

So now, what's the probability $P[0]$ we'll see $0$ when we measure just the first qubit? Well the answer is, it's the same as the probability we see $0$ on the first qubit if we measure both qubits. So what is that? Well, there are only two cases in which we'd see a $0$, the first two in the formula above so it would be $P[0] = |\alpha_{00}|^2 + |\alpha_{01}|^2$ And here's the more interesting question: what's the new state? And the answer is very elegant. What you do is cross out all those parts of the superposition that are not consistent with the answer you got. 

So you got an answer of 0, so you cross off these two possibilities. This is what the new state should look like, what's left over. Except that it's not normalized, so you just normalize it. So you'd get this state, which is not normalized, but now to normalize it, you divide by square root of $|\alpha_{00}|^2+|\alpha_{01}|^2$.
Which is also the square root of the probability with which you saw this
particular outcome.

OK, so let's just carry this through for this example.

$$|\psi\rangle = \Big(\frac{1}{2}+\frac{i}{2}\Big)|00\rangle + \frac{1}{2}|01\rangle + \frac{i}{2}|11\rangle$$

What's the probability you'll see 0? Well, it's 

$$P[0] = \bigg|\frac{1}{2}+\frac{i}{2}\bigg|^2 + \bigg(\frac{1}{2}\bigg)^2= \frac{1}{2} + \frac{1}{4} = \frac{3}{4}$$

So the probability you'll see $0$ is exactly $\frac{3}{4}$.
Now what's the new state? And you get the new state by crossing off everything that is inconsistent, so you get rid of the $|11\rangle$, and now we need to re-normalize. So we take this (resulting) state and we re-normalize it by dividing by the square root of this probability of 0:

$$|\psi'\rangle = \frac{\big(\frac{1}{2} + \frac{i}{2}\big)|00\rangle + \frac{1}{2}|01\rangle}{\sqrt{\frac{3}{4}}}$$
