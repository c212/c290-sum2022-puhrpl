**Problem 7.** In $\mathbb{C}^2$, how many real unit vectors are there whose projection onto $|1\rangle$ has length $\frac{\sqrt{3}}{2}$?

**Solution:** let's start by trying to determine the kind of thing(s) that we're looking for. 

We're looking for a vector $v = \begin{pmatrix} a \\ b \end{pmatrix}$ such that $a, b \in \mathbb{R}$ and $a^2+b^2=1$.

How do we determine the projection of a vector onto another? 

We take the inner product and divide by the length of the other vector. 

So $<1|v> = \begin{pmatrix} 0 & 1 \end{pmatrix} \cdot \begin{pmatrix} a \\ b \end{pmatrix} = b$ and the magnitude of $|1\rangle$ is 1 (unit vector). 

That means: $|b| = \frac{\sqrt{3}}{2}$ which gives the following values for $b = \pm \frac{\sqrt{3}}{2}$. 

What about $a$? We know that $a = \pm \sqrt{1 - b^2} = \pm \sqrt{1 - \frac{3}{4}} = \pm \frac{1}{2}$

So we have four vectors: $\begin{pmatrix}\pm\frac{1}{2} \\ \pm\frac{\sqrt{3}}{2}\end{pmatrix} = \pm \frac{1}{2} \pm \frac{\sqrt{3}}{2} i$

**Problem 9.** In the double-slit experiment, consider the point $x$ at the middle of the final (detector) screen which is equidistant from the two slits. Suppose the intensity at that point is $1$ when either slit is open. Now for each of the three cases (a) bullet (b) wave (c) quantum mechanics (photons or electrons) calculate the intensity at the same point when both slits are open.

**Solution:** In the case (a) bullets are discrete, the probability of arrival is $P_{12}(x) = P_1(x) + P_2(x)$ and there is no interference. We know that $P_{1}(x)=1$ and $P_{2}(x)=1$ as given. So we are asked to calculate $P_{12}(x)= 1 + 1 = 2$. In the case (b) water waves, they are continuous, the intensity is $I_{12}(x) \ne I_1(x) + I_2(x)$ and (or, rather, because) there is interference. We also have $I_1(x) = h_1^2(x)$ and $I_2(x) = h_2^2(x)$ as well as $h_{12}(x) = h_1(x) + h_2(x)$ (so heights add up but the intensities do not). We are asked to determine $I_{12}(x) = h_{12}^2(x)$ and we calculate as follows $I_{12}(x) = (h_1(x) + h_2(x))^2 = (1 + 1)^2 = 2^2 = 4$. 

Now we can predict that the answer for (c) is going to be similar since the math is almost the same. So when we deal with photons (or electrons) they are discrete, the probability of arrival is $P_{12}(x) \ne P_1(x) + P_2(x)$ and here we have a complex amplitude instead of the height of a wave. The formulas are: $P_1(x)=|a_1(x)|^2 = 1$, $P_2(x) = |a_2(x)|^2 = 1$ (these are given) and we need to find $P_{12}(x) = |a_{12}(x)|^2$. But we remember that like in the case of water waves even if the probabilities do not add up, the complex amplitudes do. Thus we write $a_{12}(x) = |a_1(x)|+|a_2(x)| = 1 + 1 = 2$ which means that $P_{12}(x) = |a_{12}(x)|^2=(1+1)^2 = 2^2=4$.

Measurement II: Measure in the standard basis and interpret $0$ as $|u\rangle$ and $1$ as $|v\rangle$. 

The probability of success is defined as $\frac{1}{2}p_u+\frac{1}{2}p_v$, where 

$$p_u = \textrm{Pr[we guess} |u\rangle \textrm{given that the qubit was in state}|u\rangle\textrm{]}$$ and

$$\sin{\theta} \approx \theta \\
